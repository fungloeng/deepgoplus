```python
# ==================== 可编辑参数 ====================
DO_RENAME = True          # 先设为 False 预览；确认后改为 True 正式改名
FOLDER   = '.'             # 要处理的目录，'.' 表示 notebook 所在目录
OLD_PART = '_run2345'     # 需要被替换的字符串
NEW_PART = '_run2'    # 替换后的字符串
# ====================================================

import os
from pathlib import Path

folder = Path(FOLDER).resolve()
targets = [p for p in folder.iterdir() if p.is_file() and OLD_PART in p.name]

if not targets:
    print("目录下没有包含 '{}' 的文件，无需改名。".format(OLD_PART))
else:
    # 构造 (旧路径, 新名字)
    rename_map = {f: f.name.replace(OLD_PART, NEW_PART, 1) for f in targets}

    # 简单冲突检测
    if len(set(rename_map.values())) != len(rename_map):
        print("⚠️  改名后会出现重名冲突，请检查！")
    else:
        print("======== 预览 ========")
        for old_path, new_name in rename_map.items():
            print(f"{old_path.name}  ->  {new_name}")
        print("=====================")

        if DO_RENAME:
            for old_path, new_name in rename_map.items():
                new_path = old_path.with_name(new_name)
                old_path.rename(new_path)
                print(f"✅  renamed: {old_path.name}  ->  {new_name}")
            print("全部改名完成！")
        else:
            print("\n已列出预览。确认无误后，把 DO_RENAME 设为 True 再运行一次即可正式改名。")
```
